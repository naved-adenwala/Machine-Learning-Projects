```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import datetime
from pandas_datareader import data
import seaborn as sns 
%matplotlib inline
```

```python
companies_dict = {
    'Amazon':'AMZN',
    'Apple':'AAPL',
    'Walgreen':'WBA',
    'Northrop Grumman':'NOC',
    'Boeing':'BA',
    'Lockheed Martin':'LMT',
    'McDonalds':'MCD',
    'Intel':'INTC',
    'IBM':'IBM',
    'Texas Instruments':'TXN',
    'MasterCard':'MA',
    'Microsoft':'MSFT',
    'General Electrics':'GE',
    'American Express':'AXP',
    'Pepsi':'PEP',
    'Coca Cola':'KO',
    'Johnson & Johnson':'JNJ',
    'Toyota':'TM',
    'Honda':'HMC',
    'Exxon':'XOM',
    'Chevron':'CVX',
    'Valero Energy':'VLO',
    'Ford':'F',
    'Bank of America':'BAC'}
```

```python
start = "2015-04-25"
end =  "2020-04-25"

df = data.DataReader(list(companies_dict.values()),data_source="yahoo",start=start,end=end)
```

```python
stock_open = np.array(df["Open"]).T
stock_close = np.array(df["Close"]).T

change = stock_open - stock_close
```

```python
from sklearn.preprocessing import Normalizer
normalizer = Normalizer() # Define a Normalizer
norm_movements = normalizer.fit_transform(change) # Fit and transform
```

```python
print(norm_movements.min())
print(norm_movements.max())
print(norm_movements.mean())
```

```python
plt.figure(figsize = (20,8)) 
ax1 = plt.subplot(1,2,1)
plt.title('Company:Amazon',fontsize = 20)
plt.xticks(fontsize = 18)
plt.yticks(fontsize = 20)
plt.xlabel('Date',fontsize = 20)
plt.ylabel('Movement',fontsize = 20)
plt.plot(norm_movements[0]) 
plt.subplot(1,2,2,sharey = ax1)
plt.title('Company:Apple',fontsize = 20)
plt.xticks(fontsize = 18)
plt.yticks(fontsize = 20)
plt.xlabel('Date',fontsize = 20)
plt.ylabel('Movement',fontsize = 20)
plt.plot(norm_movements[1])
```

```python
from sklearn.pipeline import make_pipeline
from sklearn.cluster import KMeans
```

```python
# Define a normalizer
normalizer = Normalizer()
# Create Kmeans model
kmeans = KMeans(n_clusters = 10,max_iter = 1000)
# Make a pipeline chaining normalizer and kmeans
pipeline = make_pipeline(normalizer,kmeans)
# Fit pipeline to daily stock movements
pipeline.fit(change)
predictions = pipeline.predict(change)
pd.DataFrame({"Labels":predictions,"Company":list(companies_dict)}).sort_values(by="Labels",axis=0)
```

```python
from sklearn.decomposition import PCA
# Define a normalizer
normalizer = Normalizer()
#PCA model
pca = PCA(n_components=2)

# Create Kmeans model
kmeans = KMeans(n_clusters = 10,max_iter = 1000)
# Make a pipeline chaining normalizer and kmeans
pipeline = make_pipeline(normalizer,pca,kmeans)
# Fit pipeline to daily stock movements
pipeline.fit(change)
predictions = pipeline.predict(change)
```

```python
pd.DataFrame({"Labels":predictions,"Company":list(companies_dict)}).sort_values(by="Labels",axis=0)
```
