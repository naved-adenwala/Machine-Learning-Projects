```python
total_scase = df.sum()
```

```python
total_scase = total_scase.drop(["sno","state_name","state_code"])
```

```python
total_scase
```

```python
d = total_scase[["active","positive","cured","death"]]

plt.figure(figsize=(5,5))
plt.pie(d,explode=[0,0,0,2],autopct="%1.1f%%",startangle=15,labels=["active","positive","cured","death"],colors=["red","lightblue","green","purple"],shadow=True)
```

```python
#for new cases
d = total_scase[["new_active","new_positive","new_cured","new_death"]]

plt.figure(figsize=(5,5))
plt.pie(d,explode=[0,0,0,2],autopct="%1.1f%%",startangle=15,labels=["new_active","new_positive","new_cured","new_death"],colors=["red","lightblue","green","purple"],shadow=True)
```

```python
#all cases
#for new cases
d = total_scase[["active","positive","cured","death","new_active","new_positive","new_cured","new_death"]]

plt.figure(figsize=(5,5))
plt.pie(d,explode=[0,0,0,0,0,0,0,2],autopct="%1.1f%%",startangle=15,labels=["active","positive","cured","death","new_active","new_positive","new_cured","new_death"],shadow=True)
```
